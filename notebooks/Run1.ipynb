```python
import pandas as pd
import torch
import matplotlib.pyplot as plt
```

```python
map_str = {
    'avg_loss': 'avg loss',
    'dead_features': 'avg dead neurons'
}

def plotter(logs, log_types, model_prefix, save_path="plot.png"):
    num_logs = len(log_types)
    fig, axs = plt.subplots(1, num_logs, figsize=(6 * num_logs, 5), squeeze=False)
    
    for idx, log_type in enumerate(log_types):
        ax = axs[0][idx]
        for model, model_logs in logs.items():
            if not model.startswith(model_prefix):
                continue
            values = model_logs[log_type]
            epochs = list(range(1, len(values) + 1))
            ax.plot(epochs, values)
                
        ax.set_xlabel("Epoch")
        ax.set_ylabel(map_str[log_type])
        ax.grid(True)
        ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.2f}'))
        # legend removed
    
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.close(fig)
```

```python
logs = torch.load('../runs/seed1/sae_train_logs/topk_sae/top_k_sae_results_reanim.pt')
plotter(logs, ['avg_loss'], 'reanim_topk', 'topk_avg_loss.png')
```

```python
def plot_single_penalty_heatmap(all_concept_paths, group, penalty, thresh_type='top_k_10', save_path=None):
    """
    Zoom in on a single group's heatmap for a specific penalty and optionally save as PNG.
    
    Args:
        all_concept_paths: Dictionary containing all SAE results
        group: Tuple (has_reanim, dict_size)
        penalty: Penalty value to plot
        thresh_type: Threshold type to use
        save_path: File path to save the figure (e.g., 'heatmap.png'). If None, figure is not saved.
    """
    has_reanim, dict_size = group
    # Find the matching SAE
    for sae_name, results_by_thresh in all_concept_paths.items():
        h_reanim, d_size, p = parse_sae_name(sae_name)
        if h_reanim == has_reanim and d_size == dict_size and p == penalty:
            if thresh_type not in results_by_thresh:
                print(f"Threshold type '{thresh_type}' not found for SAE {sae_name}")
                return
            data = results_by_thresh[thresh_type]
            if 'concept_paths' not in data or 'summary' not in data:
                print(f"No concept paths or summary found for SAE {sae_name}")
                return
            
            concept_paths = data['concept_paths']
            summary = data['summary']
            
            # Calculate neuron usage matrix
            num_neurons = summary.get('n_neurons', 256)
            num_classes = 10
            neuron_usage_matrix = np.zeros((num_classes, num_neurons))
            samples_per_class = defaultdict(int)
            
            for path in concept_paths:
                label = path.get('label')
                if label is not None and 0 <= label < num_classes:
                    samples_per_class[label] += 1
                    for neuron_idx in path['active_neurons']:
                        if 0 <= neuron_idx < num_neurons:
                            neuron_usage_matrix[label, neuron_idx] += 1
            
            for i in range(num_classes):
                if samples_per_class[i] > 0:
                    neuron_usage_matrix[i, :] /= samples_per_class[i]
            
            # Plot with full axes and colorbar, remove x-axis labels
            plt.figure(figsize=(12, 6))
            sns.heatmap(neuron_usage_matrix, cmap='viridis', 
                        xticklabels=False, yticklabels=range(num_classes), cbar=True)
            reanim_label = "Reanim" if has_reanim else "Vanilla"
            plt.title(f'{reanim_label}, Dict Size: {dict_size}, Penalty: {penalty}, Thresh: {thresh_type}', fontsize=14)
            plt.ylabel('Digit')
            plt.xlabel('')  # Remove x-axis label
            
            # Save figure if save_path is provided
            if save_path:
                plt.savefig(save_path, dpi=300, bbox_inches='tight')
                print(f"Heatmap saved to {save_path}")
            
            plt.show()
            return
    
    print(f"No matching SAE found for group {group} with penalty {penalty}")
```
